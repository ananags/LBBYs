```python
def parse_currency(value):
    if pd.isna(value):
        return None
    return float(str(value).replace('$', '').replace(',', ''))
data.head()
data['DisbursementGross'] = data['DisbursementGross'].apply(parse_currency)
data.head()
```

```python
import pandas as pd

df = pd.read_csv('../3_modelado/3.3_modelo_secreto/train_limpio.csv')

# 1. Eliminar columnas irrelevantes
cols_to_drop = ['id', 'LoanNr_ChkDgt', 'Name']
df.drop(columns=cols_to_drop, inplace=True)

# 2. Reemplazar todos los NaNs por 'Unknown', sin suponer valores
df = df.fillna('Unknown')

# 3. Conversión de fechas: convertir pero mantener Unknown si no se puede parsear
for date_col in ['ApprovalDate', 'DisbursementDate']:
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    df[date_col] = df[date_col].fillna('Unknown')

# 4. Extraer año y mes si la fecha es válida, si no poner 'Unknown'
df['ApprovalYear'] = df['ApprovalDate'].apply(lambda x: x.year if x != 'Unknown' else 'Unknown')
df['ApprovalMonth'] = df['ApprovalDate'].apply(lambda x: x.month if x != 'Unknown' else 'Unknown')

# 5. Limpiar y convertir BalanceGross (monetario)
df['BalanceGross'] = df['BalanceGross'].astype(str).replace('[\$,]', '', regex=True)
df['BalanceGross'] = df['BalanceGross'].apply(lambda x: float(x) if x.replace('.', '', 1).isdigit() else 'Unknown')

# 6. Mapear binarios si son Y/N o Yes/No, si no están especificados → '2'
def map_binary(val):
    if val in ['Y', 'Yes']:
        return 1
    elif val in ['N', 'No']:
        return 0
    else:
        return 2

df['RevLineCr'] = df['RevLineCr'].apply(map_binary)
df['LowDoc'] = df['LowDoc'].apply(map_binary)
df['IsFranchise'] = df['IsFranchise'].apply(map_binary)

# 7. Convertir variables categóricas (CatBoost las puede manejar como 'category')
categorical_columns = ['City', 'State', 'Bank', 'BankState']
for col in categorical_columns:
    df[col] = df[col].astype(str) 
```

```python
df.to_csv('train_limpio_v2.csv')
```

## Conclusiones

### Limpieza de columnas categóricas

RevLineCr y LowDoc contenían valores anómalos como '0', 'T', o símbolos ('), que fueron corregidos y estandarizados a 'Y', 'N', o 'NA'.

En LowDoc se detectaron valores erróneos ('A', 'R', 'S', 'C') que fueron considerados como información faltante y recodificados como 'NA'.

Se creó una nueva columna IsFranchise, que binariza FranchiseCode: 'No' si es '0' o '1', 'Sí' en caso contrario
